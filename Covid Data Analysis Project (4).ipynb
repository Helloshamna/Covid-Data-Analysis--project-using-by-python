# Active cases


```python
covidindia_df['Active_Cases']=covidindia_df['Confirmed']-(covidindia_df['Cured']+covidindia_df['Deaths'])
covidindia_df.tail()
```

```python
statewise=pd.pivot_table(covidindia_df,values=['Confirmed','Deaths','Cured'],  index ='State/UnionTerritory',aggfunc=max)
```

```python
statewise["Recovery Rate"]=statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise["Mortality Rate"]=statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise=statewise.sort_values(by="Confirmed",ascending=False)
```

```python
statewise.style.background_gradient(cmap= "cubehelix")
```

# Top 10active cases states

```python
top_10_active_cases = (
    covidindia_df
    .groupby('State/UnionTerritory', as_index=False)[['Active_Cases', 'Date']]
    .max()
    .sort_values(by='Active_Cases', ascending=False)
    .reset_index(drop=True)
)

top_10_active_cases.head(10)  # Display top 10 rows
```

```python
fig=plt.figure(figsize=(16,9))
```

```python

top_10_active_cases = (
    covidindia_df
    .groupby('State/UnionTerritory', as_index=False)[['Active_Cases', 'Date']]
    .max()
    .sort_values(by='Active_Cases', ascending=False)
    .reset_index(drop=True)
)

fig=plt.figure(figsize=(16,9))
plt.title("Top 10 States with most active cases in india",size=25)
ax=sns.barplot(data=top_10_active_cases.iloc[:10], y= "Active_Cases",x= "State/UnionTerritory",linewidth=2, edgecolor='red')
plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()

```

# Top  states  with  hightst deaths

```python
top_10_deaths=covidindia_df.groupby('State/UnionTerritory', as_index=False)[['Deaths', 'Date']].max().sort_values(by='Deaths', ascending=False).reset_index(drop=True)

fig=plt.figure(figsize=(20,6))
plt.title("Top 10 States with most Death",size=20)
ax=sns.barplot(data=top_10_deaths.iloc[:12],y="Deaths",x= "State/UnionTerritory",linewidth =5,edgecolor="k")
plt.xlabel("States")
plt.ylabel("Total Death Cases")
plt.show()
```

# Remove rows where sate= India

```python
vaccine=covidvacin_df[covidvacin_df.State!='India']
vaccine
```

```python
vaccine.rename(columns={"Total Individuals Vaccinated":"Total"},inplace= True)
vaccine.head()
```

# Most vaccinated State

```python
max_vac=vaccine.groupby("State")['Total'].sum().to_frame("Total")
max_vac=max_vac.sort_values('Total',ascending=False)[:5]
max_vac
```

```python
fig=plt.figure(figsize=(10,6))
plt.title("Top 5 Vaccinated States in India",size=20)
ax=sns.barplot(data=max_vac.iloc[:10],y=max_vac.Total,x= max_vac.index,linewidth =2,edgecolor="r")
plt.xlabel("States")
plt.ylabel("Vaccination")
plt.show()
```
